```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/insurance - insurance.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
df['sex']=lb.fit_transform(df['sex'])
df['smoker']=lb.fit_transform(df['smoker'])
df['region']=lb.fit_transform(df['region'])
```

```python
x=df.drop(columns=['charges'])
y=df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
#ridge regression model
ridgeReg=Ridge(alpha=10)
ridgeReg.fit(x_train,y_train)
#train and test score for ridge reggression
train_score_ridge=ridgeReg.score(x_train,y_train)
test_score_ridge=ridgeReg.score(x_test,y_test)

```

```python
print('the train score bridge mode is{}',format(train_score_ridge))
print('the test score bridge mode is{}',format(test_score_ridge))
```

```python
#lasso
```

```python
#lasso regression model
lasso=Lasso(alpha=10)
lasso.fit(x_train,y_train)
#train and test score for lasso reggression
train_score_lasso=lasso.score(x_train,y_train)
test_score_lasso=lasso.score(x_test,y_test)
```
