```python
df=pd.read_csv("/content/cars - cars.csv")
```

```python
df
```

```python
df['km_driven'].describe()
```

```python
sns.boxplot(x=df['km_driven'])
```

```python
percentile25=df['km_driven'].quantile(0.25)
percentile75=df['km_driven'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
IQR=percentile75-percentile25
```

```python
IQR
```

```python
upper_limit=percentile75+IQR*0.75
upper_limit
```

```python
lower_limit=percentile25+IQR*0.25
lower_limit
```

```python
df[df['km_driven']>upper_limit]
```

```python
df[df['km_driven']<lower_limit]
```

```python
newdf=df[df['km_driven']<upper_limit]
```

```python
plt.figure(figsize=(15,7))
plt.subplot(222)
sns.boxplot(x=df['km_driven'])
plt.subplot(224)
sns.boxplot(x=newdf['km_driven'])
plt.show()
```

```python
new_df_cap=df.copy()
```

```python
new_df_cap['km_driven']=np.where(

    new_df_cap['km_driven']>upper_limit,
    upper_limit,
    np.where(
    new_df_cap['km_driven']<lower_limit,
    lower_limit,
    new_df_cap['km_driven']
))
```

```python
plt.figure(figsize=(15,7))
plt.subplot(222)
sns.boxplot(x=df['km_driven'])
plt.subplot(224)
sns.boxplot(x=new_df_cap['km_driven'])
plt.show()
```
