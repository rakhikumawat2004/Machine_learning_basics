```python
# function transformer
#The function transformer is a tool in scikit learn a popular python libraray
#for machine learning that allows tou to apply a spcified function to the input
#data . The function transformer can be useful for performing custom
# transformations of input data in a machine learning pipeline
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
```

```python
#create a data set
x=np.array([[1,2],[3,4]])
```

```python
# define the transformation function
log_transform=FunctionTransformer(np.log1p)
```

```python
#apply the transformation to the the data set
x_transformed=log_transform.transform(x)
```

```python
#view the transform data
print(x_transformed)
```

```python
#create a data set
x=np.array([[1,2],[3,4]])
```

```python
#define a custom feature engineering function
def squ(x):
  return np.hstack((x,x**2))
```

```python
#create a function transformer to apply the custom function
custom_transformer=FunctionTransformer(squ)
```

```python
#apply the transformer to the input data
x_transformed=custom_transformer.transform(x)
```

```python
#view the transformed data
print(x_transformed)
```

```python
#create a data set
x=np.array([[1,2],[3,4]])
```

```python
#define custom scaling function
def my_scaling(x):
  return x/np.max(x)
```

```python
#create a function transformer to apply the custom function
custom_transformer=FunctionTransformer(my_scaling)
```

```python
#apply the transformer to the data set
x_transformed=custom_transformer.transform(x)
```

```python
#view the transformed data
print(x_transformed)
```

```python
#create a data set with missing values
x=np.array([[1,2],[3,np.nan]])
```

```python
#define custom cleaning function
def my_cleaning(x):
  x[np.isnan(x)]=0
  return x
```

```python
#create a function transformer to apply the custom function
custom_transformer=FunctionTransformer(my_cleaning)
```

```python
#apply the transformer to the input data set
x_transformed=custom_transformer.transform(x)
```

```python
#view the transformed data
print(x_transformed)
```
