```python
df=pd.read_csv("/content/titanic - titanic.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Age'].describe()
```

```python
sns.boxplot(x=df['Age'])
```

```python
percentile25=df['Age'].quantile(0.25)
```

```python
percentile75=df['Age'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
IQR=percentile75-percentile25
```

```python
IQR
```

```python
upper_limit=percentile75+IQR*0.75
```

```python
upper_limit
```

```python
lower_limit=percentile25+IQR*0.25
```

```python
lower_limit
```

```python
df[df['Age']>upper_limit]
```

```python
df[df['Age']<lower_limit]
```

```python
newdf=df[df['Age']<upper_limit]
newdf=df[df['Age']>lower_limit]
```

```python
newdf
```

```python
plt.figure(figsize=(15,7))
plt.subplot(222)
sns.boxplot(x=df['Age'])
plt.subplot(224)
sns.boxplot(x=newdf['Age'])
plt.show()
```

```python
new_df_cap=df.copy()
```

```python
new_df_cap['Age']=np.where(

    new_df_cap['Age']>upper_limit,
    upper_limit,
    np.where(
    new_df_cap['Age']<lower_limit,
    lower_limit,
    new_df_cap['Age']
))
```

```python
plt.figure(figsize=(15,7))
plt.subplot(222)
sns.boxplot(x=df['Age'])
plt.subplot(224)
sns.boxplot(x=new_df_cap['Age'])
plt.show()
```
